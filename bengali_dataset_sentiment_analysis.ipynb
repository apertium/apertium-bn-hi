Combining the individual dataframes to form a single dataframe

```python
#X=np.array(df.Word.values)
X=df.Word.values
```

**Encoding indivisual character**s

```python
d={}
string="কখগঘঙচছজঝঞটঠডঢণতথদধনপফবভময়ড়ঢ়যরলহশষস্অআইঈউঊািীুূঋৠঌৡএঐওঔৃৄৢৣৈেোৌৎৗঁংঃ়ঽ"
for i in range(len(string)):
  l=[0 for j in range(len(string))]
  l[i]=1
  d[string[i]]=np.array(l)
print(d)
```

```python
print(len(string))
```

Encoding the sequence of characters with the vector as determined above. The size is maximum of the number of characters in the string

```python
maxim=0
for i in range(len(X)):
    l=[]
    string=str(X[i])
    if len(string)>maxim:
      maxim=len(string)
    print(string)
    l2=string.split("_")
    l=[]
    #print(len(string)),print(string)
    for k in l2:
      for j in range(len(k)):

        if k[j]!="-":
          try:
            l.append(string.index(k[j])
          except:
            continue
    l=np.array(l)
    X[i]=l
```

```python
print(maxim)
```

```python
len(string)
```

```python
X=tf.keras.preprocessing.sequence.pad_sequences(X,maxlen=42,padding="post")
```

Determining the embedding weights

```python
embedding_weights=[]
for i in d:
  embedding_weights.append(d[i])
embedding_weights=np.array(embedding_weights)
```

```python
embedding_weights.shape
```

```python
y=df.sentiment.values
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
#X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
#X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#using LSTM
model=tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(70,70,weights=[embedding_weights]))
model.add(tf.keras.layers.CuDNNLSTM(70,return_sequences=True))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.CuDNNLSTM(70,return_sequences=False))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(3,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=20)
```

```python
model.summary()
```

```python

test_loss, test_acc = model.evaluate(X_test,y_test)

print('Test Loss: {}'.format(test_loss))
print('Test Accuracy: {}'.format(test_acc))
```

```python
#using RNN
model=tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(70,70,weights=[embedding_weights]))
model.add(tf.keras.layers.SimpleRNN(70,return_sequences=True))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.SimpleRNN(70))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(3,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=12)
```

```python

test_loss, test_acc = model.evaluate(X_test,y_test)

print('Test Loss: {}'.format(test_loss))
print('Test Accuracy: {}'.format(test_acc))
```

```python
#using CNN
model=tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(70,70,weights=[embedding_weights]))
model.add(tf.keras.layers.Conv1D(350,4))
model.add(tf.keras.layers.GlobalMaxPooling1D())
model.add(tf.keras.layers.Dense(350))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(3,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
test_loss, test_acc = model.evaluate(X_test,y_test)

print('Test Loss: {}'.format(test_loss))
print('Test Accuracy: {}'.format(test_acc))
```
